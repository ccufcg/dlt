```python
import hashlib
import secrets

def cadastrar(senha: str) -> tuple[str, str]:
    salt = secrets.token_hex(16)          # salt unico por usuario
    digest = hashlib.sha256(
        (salt + senha).encode()
    ).hexdigest()
    return salt, digest                   # armazenar no banco

def verificar(senha: str, salt: str, digest: str) -> bool:
    digest_calc = hashlib.sha256(
        (salt + senha).encode()
    ).hexdigest()
    return secrets.compare_digest(digest_calc, digest)
```

```python
salt, digest = cadastrar("minha_senha")
print(f"Salt:   {salt}")
print(f"Digest: {digest}")
print(f"Valido: {verificar('minha_senha', salt, digest)}")
```
